```python
### Using a dictionary dict to store the columns and the corresponding percentage of missing values in the column.

a=[]
dict = {}
v = 0

# This function calculates the percentage of missing values for a particular column.

def missingvalues (i):
    v =(len(appl_data[i][appl_data[i].isnull()])/len(appl_data) *100)
    return(v)

#This loop is used to create a dictionary which stores the 'column names' as dictionary keys and
#the 'missing value percentage' as dictionary values.

for a in c:
    if len(appl_data[a][appl_data[a].isnull()]) > 0:
        v = missingvalues(a)
        dict[a] = v
       

print(dict)  
```

```python
### Selecting the number of columns which have a data missing percentage equals 50% or higher in the col_drop list

col1= dict.keys()

col_drop = []
for a in col1:
    if dict[a] >= 50:
        col_drop.append(a)

print(col_drop)
    
len(col_drop)
```
