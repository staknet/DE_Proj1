```python
#let's create the country table

countryTableCreate = """CREATE TABLE IF NOT EXISTS accountscountry(
country_code VARCHAR PRIMARY KEY,
short_name VARCHAR,
table_name VARCHAR,
long_name VARCHAR,
currency_unit VARCHAR)"""
```

```python
try:
    cur.execute(countryTableCreate)
    con.commit()
    print("table created")
except psycopg2.Error as e:
    print(e)
```
